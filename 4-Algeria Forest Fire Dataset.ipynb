# Model Training

```python
df1 = pd.read_csv("Algerian_forest_fires_dataset_cleaned_dataset.csv")
```

```python
df1.head()
```

```python
df1.columns
```

```python
### drop day,month,year
df1.drop(['day','month','year'],axis = 1, inplace = True)
```

```python
df1.head()
```

```python
df1["Classes"].value_counts()
```

```python
## Encoding 
df1["Classes"] = np.where(df1["Classes"].str.contains("not fire"),0,1)
```

```python
df1.tail()
```

```python
df1["Classes"].value_counts()
```

```python
## Independent and Dependent features
X = df1.drop('FWI', axis = 1)
y = df1['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlation
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize = (12,10))
corr = X_train.corr()
sns.heatmap(corr,annot = True)
plt.show()
```

```python
def correlation(df1, threshold):
    col_corr = set()
    corr_matrix = df1.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

```python
## threshold -- Domain expertise
corr_features = correlation(X_train,0.85)
```

```python
## drop features when correlatrion is more than 0.85
X_train.drop(corr_features, axis = 1, inplace = True)
X_test.drop(corr_features, axis = 1, inplace = True)

X_train.shape, X_test.shape
```

# Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## Box plot to understand effect of Standard Scaler


```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1,)
sns.boxplot(data = X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data = X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute error", mae)
print("R2_score",score)
```

```python
plt.scatter(y_test,y_pred)
plt.show()
```
