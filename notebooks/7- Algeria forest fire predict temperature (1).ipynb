```python
df1 = pd.read_csv("Algerian_forest_fires_cleaned_dataset.csv7")
```

```python
df1.columns
```

```python
## Drop day,month,year
df1.drop(['day', 'month', 'year'],axis = 1,inplace = True)
```

```python
df1.head()
```

```python
df1["Classes"].value_counts()
```

```python
## Encoding
df1['Classes'] = np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df1.tail()
```

```python
df1['Classes'].value_counts()
```

```python
## Independent and dependent features (FWI = Fire Weather Index)
X = df1.drop('FWI',axis = 1)
y = df1['FWI']
```

```python
X.head()
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlation
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize = (12,10))
corr = X_train.corr()
sns.heatmap(corr,annot = True)
```

```python
def correlation(dataset, threshold): ## My independent feature should not be greater than than the threshold,if it is greater than threshold we can remove the feature.
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

```python
## threshold  -- Domain Expertise
corr_features = correlation(X_train,0.85)
```

```python
corr_features
```

```python
## Drop features when correlation is more than 0.85
X_train.drop(corr_features,axis = 1, inplace = True)
X_test.drop(corr_features, axis = 1, inplace = True)
X_train.shape,X_test.shape
```

## Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

# Box Plots To understand Effect of Standard Scaler

```python
plt.subplots(figsize = (15,5))
plt.subplot(1,2,1)
sns.boxplot(data = X_train)
plt.title("X_train Before Scaling")
plt.subplot(1,2,2)
sns.boxplot(data = X_train_scaled)
plt.title("X_train After Scaling")
```

```python
# After Scaling the ranges is same 
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean_absolute_error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error,r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```
